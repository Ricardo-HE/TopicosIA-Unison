### $H1$

#### 10 Datos

```python
no_datos = 5
#datos = (np.random.uniform(-1,1,5))
datos = [(random.uniform(-1,1), random.uniform(-1,1)) for _ in range(no_datos)]
h1 = [(datos[i][0] + datos[i][1])/2 for i in range(no_datos)]
H1.append(h1)

fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
x1 = np.linspace(-1,1,300)
y1 = np.sin(np.pi*x1)

ax.plot(x1, y1)
for i in range(no_datos):
    ax.hlines(h1[i], xmax=1, xmin=-1)

plt.title(r'$H1$ - 10 datos', fontsize=20)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$y$', fontsize=18)
    
plt.savefig('H1-10-2.png')
```

### $H2$


```python
H2_a = []
H2_b = []
```

#### 10 Datos

```python
no_datos = 5

#datos = (np.random.uniform(-1,1,5))
x = [(random.uniform(-1,1), random.uniform(-1,1)) for _ in range(no_datos)]
y = [(np.sin(d[0]), np.sin(d[1])) for d in x]

a = np.array([(y_[1] - y_[0])/(x_[1]-x_[0]) for x_,y_ in zip(x,y)])
b = np.array([y[i][0] - a[i]*x[i][0] for i in range(no_datos)])

H2_a.append(list(a))
H2_b.append(list(b))

fig, ax = plt.subplots()
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
x1 = np.linspace(-1,1,300)
y1 = np.sin(np.pi*x1)

ax.plot(x1, y1)
for i in range(no_datos):
    h2 = a[i]*x1 + b[i]
    ax.plot(x1, h2)

plt.title(r'$H2$ - 10 datos', fontsize=20)
plt.xlabel(r'$x$', fontsize=18)
plt.ylabel(r'$y$', fontsize=18)
```
